```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from prettytable import PrettyTable
```

```python
# Load dataset
iris = load_iris()
X = iris.data
y = iris.target

# Split once for fair comparison
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, random_state=0
)
```

```python
# Effect of max_depth
depths = [1, 2, 3, 4, 5, None]
acc_depth = []

for d in depths:
    clf = DecisionTreeClassifier(max_depth=d, random_state=0)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc_depth.append(accuracy_score(y_test, y_pred))

# Display results
t1 = PrettyTable(["Max Depth", "Accuracy"])
for d, acc in zip(depths, acc_depth):
    t1.add_row([d, round(acc, 3)])
print("Effect of max_depth")
print(t1)

# Plot
plt.plot([str(d) for d in depths], acc_depth, marker='o')
plt.title("Accuracy vs Max Depth")
plt.xlabel("Max Depth")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```

```python
# Effect of min_samples_split
splits = [2, 5, 10, 20]
acc_split = []

for s in splits:
    clf = DecisionTreeClassifier(min_samples_split=s, random_state=0)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc_split.append(accuracy_score(y_test, y_pred))

t2 = PrettyTable(["Min Samples Split", "Accuracy"])
for s, acc in zip(splits, acc_split):
    t2.add_row([s, round(acc, 3)])
print("Effect of min_samples_split")
print(t2)

plt.plot(splits, acc_split, marker='o')
plt.title("Accuracy vs Min Samples Split")
plt.xlabel("Min Samples Split")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```

```python
# Effect of min_samples_leaf
leaves = [1, 2, 5, 10]
acc_leaf = []

for l in leaves:
    clf = DecisionTreeClassifier(min_samples_leaf=l, random_state=0)
    clf.fit(X_train, y_train)
    y_pred = clf.predict(X_test)
    acc_leaf.append(accuracy_score(y_test, y_pred))

t3 = PrettyTable(["Min Samples Leaf", "Accuracy"])
for l, acc in zip(leaves, acc_leaf):
    t3.add_row([l, round(acc, 3)])
print("Effect of min_samples_leaf")
print(t3)

plt.plot(leaves, acc_leaf, marker='o')
plt.title("Accuracy vs Min Samples Leaf")
plt.xlabel("Min Samples Leaf")
plt.ylabel("Accuracy")
plt.grid(True)
plt.show()
```
